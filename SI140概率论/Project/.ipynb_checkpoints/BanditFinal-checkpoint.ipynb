### Prob.1 
Now suppose we obtain the parameters of the Bernoulli distributions from an
oracle, which are shown in the following table. Choose N = 5000 and compute
the theoretically maximized expectation of aggregate rewards over N time slots.
We call it the oracle value. Note that these parameters $\theta_j, j \in \{1, 2, 3\}$ and oracle
values are unknown to all bandit algorithms.

The **theoretically maximiazed expectation** is given as follows:
$$
    \max_{I(t),t=1,\dots,N}\mathbb{E}\left[\sum_{t=1}^N r_{I(t)}\right] = N\cdot\theta_{I^*} = 5000\times 0.7 = 3500
$$

### Prob.2
Implement classical bandit algorithms with following settings

The simulated reward is printed after the following code.

```python
def reward(theta_j):
    return 1 if np.random.rand() < theta_j else 0

# epsilon-greedy Algorithm
def EG(N, theta, eps=.5):
    total_reward = 0
    num_arm = len(theta)
    theta_hat = [0] * num_arm
    count = [0] * num_arm
    for t in range(N):
        if np.random.rand() < eps:
            I_t = np.random.randint(num_arm)
        else:
            I_t = np.argmax(theta_hat)
        count[I_t] += 1
        r = reward(theta[I_t])
        total_reward += r
        theta_hat[I_t] += (r - theta_hat[I_t]) / count[I_t]
    return total_reward

# Upper Confidence Bound Algorithm
def UCB(N, theta, c=1):
    total_reward = 0
    num_arm = len(theta)
    I = [i for i in range(num_arm)]
    count = [1] * num_arm
    theta_hat = [reward(theta[i]) for i in range(num_arm)]
    for t in range(4, N+1):
        I_t = np.argmax([theta_hat[j] + c * np.sqrt(2 * np.log(t) / count[j]) for j in range(num_arm)])
        count[I_t] += 1
        r = reward(theta[I_t])
        total_reward += r
        theta_hat[I_t] += (r - theta_hat[I_t]) / count[I_t]
    return total_reward

# Thompson Sampling Algorithm
def TS(N, theta, alpha0, beta0):
    total_reward = 0
    num_arm = len(theta)
    alpha = alpha0.copy()
    beta = beta0.copy()
    for t in range(N):
        theta_hat = [np.random.beta(alpha[i], beta[i]) for i in range(num_arm)]
        I_t = np.argmax(theta_hat)
        r = reward(theta[I_t])
        total_reward += r
        alpha[I_t] += r
        beta[I_t] += 1 - r
    return total_reward
```

```python
N = 5000
theta = [0.7, 0.5, 0.4]

# Epsilon-greedy Algorithm
epss = [0.1, 0.5, 0.9]
for eps in epss:
    print (f"Epsilon-greedy Algorithm(eps = {eps}): {EG(N, theta, eps)};")
print()

# Upper Confidence Bound Algorithm
cs = [1, 5, 10]
for c in cs:
    print (f"Upper Confidence Bound Algorithm(c = {c}): {UCB(N, theta, c)};")
print()

# Thompson Sampling Algorithm
alphas = [[1, 1, 1], [601, 401, 2]]
betas = [[1, 1, 1], [401, 601, 3]]
for i in range(2):
    print (f"Thompson Sampling Algorithm(alpha = {alphas[i]}, beta = {betas[i]}): {TS(N, theta, alphas[i], betas[i])};")
```

### Prob.5
Give your understanding of the exploration-exploitation trade-off in bandit algorithms:  

**Overview**: We focus on the effect of N on the total rewards.

**Understand1**: With too less or too much exploration, the total reward would both decrease.  
**Explanation1**:  
-> With more exploration, we get more information about what the real distribution is, and hence may have more probability to choose the right bandit to achieve the most reward.  
-> However, with more exploration, we also may waste more time on exploring what the real distribution, or even pick the wrong bandit (hence losing much during exploitation).  
-> Consider an extreme case: if we always explore with equal probability on three bandits, the total gain will be $\frac{1}{3}N(\theta_1 + \theta_2 + \theta_3)$, which is far less than the oracle value $N \max \{\theta_1, \theta_2, \theta_3\}$. What's more, if we picked the wrong bandit, the exploitation reward would decrease a lot.   

*An experiment is given as follows: we fix 5000 rolls of bandit and $\epsilon = 0.1$. However, for different cases, we apply different number of exploration and exploitation. We figure the result between rewards and number of exploration.*

```python
def EG_full_return(N, theta, eps=.5):
    total_reward = 0
    num_arm = len(theta)
    theta_hat = [0] * num_arm
    count = [0] * num_arm
    for t in range(N):
        if np.random.rand() < eps:
            I_t = np.random.randint(num_arm)
        else:
            I_t = np.argmax(theta_hat)
        count[I_t] += 1
        r = reward(theta[I_t])
        total_reward += r
        theta_hat[I_t] += (r - theta_hat[I_t]) / count[I_t]
    return total_reward, theta_hat

real_prob = [0.7, 0.5, 0.4]

total_N = 500
explorations = np.array([1, 2, 3, 4, 5, 10, 20, 30, 40, 50, 100, 200, 300, 400, 500])
exploitations = total_N - explorations
repeats = 10000
results = np.zeros([len(exploitations)])

for i in range(len(explorations)):
    for _ in range(repeats):
        temp = EG_full_return(N=explorations[i], theta=real_prob, eps=0.1)
        results[i] += temp[0]
        results[i] += max(temp[1]) * exploitations[i]
    results[i] /= repeats

print(results)
```

```python
plt.scatter(range(len(explorations)), results)
plt.xlabel("Explorations = [1, 2, 3, 4, 5, 10, 20, 30, 40, 50, 100, 200, 300, 400, 500]")
plt.ylabel("Total Rewards")
plt.legend(["Total Attempts = 500"])
plt.title("A visual-representation of the relationship between explorations and rewards \n(Taking eps-greedy algorithm as example)")
```

```python
real_prob = [0.7, 0.5, 0.4]

points_per_N = 1000
N = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000]

EG_results = np.zeros([len(N), points_per_N], dtype=np.float64)

# summon 4000 points total; each is (N, Total_reward)
for n in range(len(N)):
    for i in range(points_per_N):
        EG_results[n][i] = EG(N=N[n], theta=real_prob, eps=0.1)

# print(EG_results)

# min-max normalization; not affecting the deviriation
for row in range(EG_results.shape[0]):
    mmin = np.min(EG_results[row, :])
    mmax = np.max(EG_results[row, :])
    EG_results[row, :] = (EG_results[row, :] - mmin) / (mmax - mmin)

for n in range(len(N)):
    print(f"Case N={N[n]}\t has variance {np.var(EG_results[n, :])}")
```

## 6.
We implicitly assume the reward distribution of these three arms are independent. How about the dependent case? Can you design an algorithm to exploit such information to obtain a better result?

### Dependent Case of Bandit Learning

-> Suppose there is a dependent network in $\{X_1, X_2, X_3\}$. The key is to find the dependent graph. Once the graph is build, the dependency relationship can be solved and we may manually find a way to maximize(exploit) rewards.  

-> Suppose the network is a **DAG**(directed acyclic graph) with **Markov Property**. We may use the CL(Chow-Liu)-algorithm to build the Bayes-Graph greedily and they apply MLE to estimate the parameters.  

-> Suppose we already have K random samples (as exploration process), labelled by $\{n_i, r_i\}$ where $n \in [1, 3]$, referring to which bandit arm, $r \in \{0, 1\}$, referring to the reward of that "attempt", $i \in [1, K]$, referring to the sample ID. The algorithm is as follows:  

1. For each pair of "bandits", or random variable $X_i, X_j$, use MLE to estimate $P(X_i)$ and $P(X_i, X_j)$. For example:
   1. $P(X_s = 1) = \frac{\sum_{i = 1}^{k}\delta (n_i = s, r_i = 1)}{\sum_{i = 1}^{k}\delta (n_i = s)}, s \in [1, 3]$
   2. $P(X_s = 1, X_t = 1) = \frac{\sum_{i = 1}^{k-1}\delta (n_i = s, n_{i + 1} = t, r_i = r_{i+1} = 1)}{\sum_{i = 1}^{k-1}\delta (n_i = s, n_{i + 1} = t)}, s,t \in [1, 3]$
2. For each pair of "bandits", or random variable $X_i, X_j$, calculate mutual information. For example:
   1. $I(X_s, X_t) = \sum_{a \in [0, 1]} \sum_{b \in [0, 1]} P(X_s = a, X_t = b) \log \frac{P(X_s = a, X_t = b)}{P(X_s = a) P(X_t = b)}$
3. Build a graph with 3 nodes. For each two nodes $X_i, X_j$, their distance is defined as $I(X_s, X_t)$, i.e. the mutual information between them.
4. Find the MST(Maximum Spanning Tree) over the graph, and build a DAG(Directed acyclic graph). This is to maximize the information that can be deducted from the dependent network.

Once the graph is build, the dependent relationship is found. We may manually choose a strategy that maximizes reward.

-> Example 1: Suppose the estimated parameters are $P(X_1 = 1) = 0.7, P(X_2 = 1 \mid X_1 = 1) = 0.9, P(X_2 = 1 \mid X_1 = 0) = 0.5, P(X_3 = 1) = 0.5$, then we may manually choose this strategy: Press $X_1$ first; if it's 1, keep pressing $X_1$, else press $X_2$ and return to pressing $X_1$.

### Prob.1
One intuitive policy suggests that in each time slot we should pull the arm for
which the current expected value of θi is the largest. This policy behaves very
good in most cases. Please design simulations to check the behavior of this policy.

```python
def intuitive_method(N, theta, gamma=0.9):
    total_reward = 0
    num_arm = len(theta)
    theta_hat = [1] * num_arm
    count = [0] * num_arm
    for t in range(1, N+1):
        I_t = np.argmax(theta_hat)
        count[I_t] += 1
        r = reward(theta[I_t])
        total_reward += gamma**(t-1) * r
        theta_hat[I_t] += (r - theta_hat[I_t]) / count[I_t]
    return total_reward
```

```python
N = 50000
theta = [0.7, 0.3]
gamma = 0.9

num_trial = 200
ret = 0
for _ in range(num_trial):
    ret += intuitive_method(N, theta, gamma)
ret /= num_trial
print (f"The result of the intuitive algorithm is {ret}.")
```

The oracal value:
$$
    \max_{I(t),t=1,\dots}\mathbb{E}\left[\sum_{t=1}^{\infty} r_{I(t)}\right] = \theta_{I^*}\cdot\sum_{t=1}^{\infty}\gamma^{t} = 0.7\times\frac{1}{1-0.9}  = 7
$$
And the gap is $\approx 0.8$, which means the behavior of this policy is not bad.

### Prob.2
However, such intuitive policy is unfortunately not optimal. Please provide an
example to show why such policy is not optimal.

For example, assume $\theta_1 > \theta_2$. 
The optimal policy is pull the first arm all the time (or at least pull the first arm more).
But if we pull the first arm and get zero reward and then pull the second arm and get some reward, we will pull the second arm for the rest of time according to this policy because $\hat{\theta_1} = 0, \hat{\theta_2} > 0$ hold after the first two pulls.

### Prob.3
For the expected total reward under an optimal policy, show that the following
recurrence equation holds:

Assuming the optimal strategy at the outset is $R(\alpha_1,\beta_1,\alpha_2,\beta_2)$ with four prior parameters $\alpha_1,\beta_1,\alpha_2,\beta_2$, where the reward obtained by pulling the $i$-th arm is $R_i(\alpha_i,\beta_i)$.

Now, with two arms, after pulling an arm, there are two possibilities: success or failure. If pulling the $i$-th arm is successful, we update the parameters as $\alpha_i\rightarrow\alpha_i+1,\beta_i\rightarrow\beta_i$ and receive the corresponding reward $1+\gamma R(\alpha_i+1,\beta_i,\cdots)$. If pulling the $i$-th arm fails, we update the parameters as $\alpha_i\rightarrow\alpha_i,\beta_i\rightarrow\beta_i+1$ and do not receive any reward $0+\gamma R(\alpha_i,\beta_i+1,\cdots)$.

For the Bayesian prior \textbf{Beta} distribution, we can estimate the corresponding parameter as $\hat\theta_i=\frac{\alpha_i}{\alpha_i+\beta_i}$. Therefore, we have:


\begin{align*}
R_1(\alpha_1,\beta_1) &= \frac{\alpha_1}{\alpha_1+\beta_1}\left[1+\gamma R(\alpha_1+1,\beta_1,\alpha_2,\beta_2)\right]+\frac{\beta_1}{\alpha_1+\beta_1}\left[\gamma R(\alpha_1,\beta_1+1,\alpha_2,\beta_2)\right] \\
R_2(\alpha_2,\beta_2) &= \frac{\alpha_2}{\alpha_2+\beta_2}\left[1+\gamma R(\alpha_1,\beta_1,\alpha_2+1,\beta_2)\right]+\frac{\beta_2}{\alpha_2+\beta_2}\left[\gamma R(\alpha_1,\beta_1,\alpha_2,\beta_2+1)\right]
\end{align*}


For each selection, we aim to find the best result. Therefore, the final choice is:

$$R(\alpha_1,\beta_1,\alpha_2,\beta_2)=\max(R(\alpha_1,\beta_1),R(\alpha_2,\beta_2))$$

### Prob.4
For the above equations, how to solve it exactly or approximately?

We may solve by monte-carlo method. First, we estimate the success probability of each arm. Then we choose that one and pull it, finally update our alpha and beta according to the given formula.

```python

def monte_carlo(num_trials, true_theta, gamma, initial_alpha_beta):
    total_rewards = 0
    alpha, beta = initial_alpha_beta.copy()

    for t in range(1, num_trials + 1):
        # 估计每个臂的成功概率
        estimated_theta = [alpha[ii] / (alpha[ii] + beta[ii]) for ii in range(2)]

        # 选择预期奖励最高的臂
        chosen_arm = np.argmax(estimated_theta)
        r = reward(true_theta[chosen_arm])
        
        alpha[chosen_arm] += r
        beta[chosen_arm] += 1 - r
        total_rewards += (gamma ** (t - 1)) * r

    return total_rewards

N = 50000
initial_alpha_beta = [[1, 1], [1, 1]]

average_reward = np.mean([monte_carlo(N, theta, gamma, initial_alpha_beta) for _ in range(100)])
print (f"The result is {average_reward}.")
```

### 5.
Find the optimal policy.

The optimal strategy should be a balanced approach between exploration and exploitation. In the initial stages, the strategy should lean towards exploration, meaning trying different arms to more accurately estimate their success probabilities. As time progresses and data accumulates, the strategy should gradually shift towards exploitation, i.e., more frequently choosing the arm with the currently highest estimated success probability. This corresponds to using the recursive formula from the third question, selecting the arm with the maximum current expected value.
