### Prob.1 
Now suppose we obtain the parameters of the Bernoulli distributions from an
oracle, which are shown in the following table. Choose N = 5000 and compute
the theoretically maximized expectation of aggregate rewards over N time slots.
We call it the oracle value. Note that these parameters $\theta_j, j \in \{1, 2, 3\}$ and oracle
values are unknown to all bandit algorithms.


Solution:
According to **theoretically maximiazed expectation**
$$  \max_{I(t),t=1,\dots,N}\mathbb{E}\left[\sum_{t=1}^N r_{I(t)}\right]=N\cdot\theta_{I^*}=5000*0.7=3500 $$

### Prob.2
Implement classical bandit algorithms with following settings


Solution:
We assume that the reward generated by the machine is only affected by each $\theta_j$,thus we use the oracle data to obtain total reward, and in our implementation, the value of oracle doesn't affect the choice of arm, which is only used in reward counting.

```python
import numpy as np
import matplotlib.pyplot as plt
#from scipy.stats import beta

def reward(theta_j):
    if np.random.rand()<theta_j:
        return 1
    else:
        return 0
N=5000
oracle=[0.7,0.5,0.4]
def epsilon_greedy(theta,N,eps=0.5):
    total=0
    num=len(theta)
    theta_head=[0]*num
    count = [0] * num
    for t in range(N):
        if np.random.rand() >= eps:#生成（0，1）随机数
            I_t = np.argmax(theta_head)
        else:
            I_t = np.random.randint(num)
        count[I_t] += 1
        r = reward(theta[I_t])
        theta_head[I_t] += (r - theta_head[I_t]) / count[I_t]
        total+= r
        
    return total
epsil=[0.3,0.6,0.9]
for eps in epsil:
    reward1=epsilon_greedy( oracle,N, eps)
    
    print (f"Epsilon-greedy Algorithm with eps = {eps}:{reward1} ;")
    
print()
# Upper Confidence Bound Algorithm
def UCB(theta,N, c=1):
    total = 0
    num = len(theta)
    count = [1] * num
    theta_head = [reward(theta[i]) for i in range(num)]
    for t in range(4, N+1):
        I_t = np.argmax([theta_head[j] + c * np.sqrt(2 * np.log(t) / count[j]) for j in range(num)])
        count[I_t] += 1
        r = reward(theta[I_t])
        total += r
        theta_head[I_t] += (r - theta_head[I_t]) / count[I_t]
        
    return total
C = [3, 6, 9]
for c in C:
    reward2=UCB(oracle,N, c)
    print (f"Upper Confidence Bound Algorithm with c = {c}: {reward2};")
print()
def Thompson(theta,N,  alpha0, beta0):
    total= 0
    num = len(theta)
    alpha = alpha0.copy()
    beta = beta0.copy()
    for t in range(N):

        I_t = np.argmax([np.random.beta(alpha[i], beta[i]) for i in range(num)])
        r = reward(theta[I_t])
        total+= r
        alpha[I_t] += r
        beta[I_t] += 1 - r
    return total

    


alphas = [[1, 1, 1], [601, 401, 2]]
betas = [[1, 1, 1], [401, 601, 3]]
for i in range(2):
    reward3=Thompson(oracle,N,  alphas[i], betas[i])
    print (f"Thompson Sampling Algorithm with alpha = {alphas[i]}, beta = {betas[i]}: {reward3};")

```

### Pro.5

Give your understanding of the exploration-exploitation trade-off in bandit algorithms:  

**Overview**: We focus on the effect of N on the total rewards.

**Understand**: With exploration out of a moderate range, the total reward may not increase like before and even would decrease.  
#### Explanation:  
Undoubtedly, With more exploration, we can get more information about the real distribution, and hence may have more probability to choose the right bandit to achieve the most reward.

However, with more exploration, we also may waste more time on exploring the real distribution, or even pick the wrong bandit because of the overfitting(hence losing much during exploitation).  

For example, consider an extreme case: if we always explore with equal probability on three bandits, the total gain will be $\frac{1}{3}N(\theta_1 + \theta_2 + \theta_3)$, which is far less than the oracle value $N \max \{\theta_1, \theta_2, \theta_3\}$. What's more, as what mentioned before, the wrong bandit will even cut reward.   

*An experiment is given as follows: we fix 5000 rolls of bandit and $\epsilon = 0.2$. However, for different cases, we apply different number of exploration and exploitation. We figure the result between rewards and number of exploration.*

```python
import numpy as np

def get_reward(theta_j):
    return 1 if np.random.rand() < theta_j else 0

def epsilon_greedy(N, probabilities, epsilon=0.5):
    total_reward = 0
    num_arms = len(probabilities)
    estimated_rewards = [0] * num_arms
    counts = [0] * num_arms
    
    for t in range(N):
        if np.random.rand() < epsilon:
            chosen_arm = np.random.randint(num_arms)  # exploration
        else:
            chosen_arm = np.argmax(estimated_rewards)  # exploitation
        
        counts[chosen_arm] += 1
        reward = get_reward(probabilities[chosen_arm])  # get the reward of the current bandit
        total_reward += reward
        
        # update estimated rewards
        estimated_rewards[chosen_arm] += (reward - estimated_rewards[chosen_arm]) / counts[chosen_arm]
    
    return total_reward, estimated_rewards

# the real probability of bandit
true_probabilities = [0.8, 0.5, 0.3]

total_rolls = 800
exploration_counts = np.array([1, 2, 4, 6, 8, 10, 20, 40, 60, 80, 100, 200, 400, 600, 800])
exploitation_counts = total_rolls - exploration_counts
num_repeats = 1000
average_rewards = np.zeros(len(exploitation_counts))

# make test
for i in range(len(exploration_counts)):
    for _ in range(num_repeats):
        total_reward, estimated_rewards = epsilon_greedy(N=exploration_counts[i], probabilities=true_probabilities, epsilon=0.2)
        average_rewards[i] += total_reward + max(estimated_rewards) * exploitation_counts[i]
    average_rewards[i] /= num_repeats

print(average_rewards)
```

```python
import matplotlib.pyplot as plt
plt.scatter(range(len(exploitation_counts)), average_rewards)
plt.xlabel("Explorations = [1, 2, 4, 6, 8, 10, 20, 40, 60, 80, 100, 200, 400, 600, 800]")
plt.ylabel("Total Rewards")
plt.legend(["Total Attempts = 800"])
plt.title("A visual-representation of the relationship between explorations and rewards \n(Taking eps-greedy algorithm as example)")
```

```python
true_probabilities = [0.8, 0.5, 0.3]

points_per_N = 1000
N = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000]

EG_results = np.zeros([len(N), points_per_N], dtype=np.float64)

# collect the sum of reward of each N
for n in range(len(N)):
    for i in range(points_per_N):
        total_reward, _ = epsilon_greedy(N=N[n], probabilities=true_probabilities, epsilon=0.2)
        EG_results[n][i] = total_reward

# min-max normalization
for row in range(EG_results.shape[0]):
    mmin = np.min(EG_results[row, :])
    mmax = np.max(EG_results[row, :])
    EG_results[row, :] = (EG_results[row, :] - mmin) / (mmax - mmin)

# Var
for n in range(len(N)):
    print(f"Case N={N[n]}\t has variance {np.var(EG_results[n, :])}")
```

### Pro.6
We implicitly assume the reward distribution of these three arms are independent. How about the dependent case? Can you design an algorithm to exploit such information to obtain a better result?

### Dependent Case of Bandit Learning

-> Suppose there is a dependent network in $\{X_1, X_2, X_3\}$. The key is to find the dependent graph. Once the graph is build, the dependency relationship can be solved and we may manually find a way to maximize(exploit) rewards.  

-> Suppose the network is a **DAG**(directed acyclic graph) with **Markov Property**. We may use the CL(Chow-Liu)-algorithm to build the Bayes-Graph greedily and they apply MLE to estimate the parameters.  

-> Suppose we already have K random samples (as exploration process), labelled by $\{n_i, r_i\}$ where $n \in [1, 3]$, referring to which bandit arm, $r \in \{0, 1\}$, referring to the reward of that "attempt", $i \in [1, K]$, referring to the sample ID. The algorithm is as follows:  

1. For each pair of "bandits", or random variable $X_i, X_j$, use MLE to estimate $P(X_i)$ and $P(X_i, X_j)$. For example:
   1. $P(X_s = 1) = \frac{\sum_{i = 1}^{k}\delta (n_i = s, r_i = 1)}{\sum_{i = 1}^{k}\delta (n_i = s)}, s \in [1, 3]$
   2. $P(X_s = 1, X_t = 1) = \frac{\sum_{i = 1}^{k-1}\delta (n_i = s, n_{i + 1} = t, r_i = r_{i+1} = 1)}{\sum_{i = 1}^{k-1}\delta (n_i = s, n_{i + 1} = t)}, s,t \in [1, 3]$
2. For each pair of "bandits", or random variable $X_i, X_j$, calculate mutual information. For example:
   1. $I(X_s, X_t) = \sum_{a \in [0, 1]} \sum_{b \in [0, 1]} P(X_s = a, X_t = b) \log \frac{P(X_s = a, X_t = b)}{P(X_s = a) P(X_t = b)}$
3. Build a graph with 3 nodes. For each two nodes $X_i, X_j$, their distance is defined as $I(X_s, X_t)$, i.e. the mutual information between them.
4. Find the MST(Maximum Spanning Tree) over the graph, and build a DAG(Directed acyclic graph). This is to maximize the information that can be deducted from the dependent network.

Once the graph is build, the dependent relationship is found. We may manually choose a strategy that maximizes reward.

-> Example 1: Suppose the estimated parameters are $P(X_1 = 1) = 0.7, P(X_2 = 1 \mid X_1 = 1) = 0.9, P(X_2 = 1 \mid X_1 = 0) = 0.5, P(X_3 = 1) = 0.5$, then we may manually choose this strategy: Press $X_1$ first; if it's 1, keep pressing $X_1$, else press $X_2$ and return to pressing $X_1$.

The following is a short code to represent the above process.

```python
import numpy as np
from sklearn.metrics import mutual_info_score
import networkx as nx

# Assume we have K samples, the format of the samples is (n_i, r_i)
samples = [(1, 1), (1, 0), (2, 1), (3, 1), (2, 0), (3, 1), (1, 1)]  # Example data
K = len(samples)

# Step 1: Calculate marginal probabilities and joint probabilities
def calculate_probabilities(samples):
    counts = {1: [0, 0], 2: [0, 0], 3: [0, 0]}  # [success count, total count]
    for n, r in samples:
        counts[n][1] += 1  # Update total count
        if r == 1:
            counts[n][0] += 1  # Update success count

    probabilities = {1: counts[1][0] / counts[1][1], 
                     2: counts[2][0] / counts[2][1],
                     3: counts[3][0] / counts[3][1]}
    
    return probabilities

# Step 2: Calculate mutual information
def calculate_mutual_information(samples):
    prob = calculate_probabilities(samples)
    mutual_info = {}
    for X_i in [1, 2, 3]:
        for X_j in [1, 2, 3]:
            if X_i < X_j:  # Avoid duplicate calculation
                joint_dist = np.zeros((2, 2))  # Joint distribution matrix
                for n1 in [0, 1]:
                    for n2 in [0, 1]:
                        joint_count = sum(1 for n, r in samples if (n == X_i and r == n1) or (n == X_j and r == n2))
                        joint_dist[n1, n2] = joint_count

                # Calculate mutual information
                mi = mutual_info_score(None, None, contingency=joint_dist)
                mutual_info[(X_i, X_j)] = mi

    return mutual_info

# Step 3: Build the graph and find the maximum spanning tree
def build_graph_and_find_mst(mutual_info):
    G = nx.Graph()
    for (X_i, X_j), mi in mutual_info.items():
        G.add_edge(X_i, X_j, weight=mi)

    mst = nx.maximum_spanning_tree(G)
    return mst

# Main program
probabilities = calculate_probabilities(samples)
mutual_info = calculate_mutual_information(samples)
mst = build_graph_and_find_mst(mutual_info)

# Print results
print("Marginal probabilities:", probabilities)
print("Mutual information:", mutual_info)
print("Maximum spanning tree:", mst.edges(data=True))
```

### Prob.3

For the expected total reward under an optimal policy, show that the following recurrence equation holds:

Solution:

Let $R(\alpha_1,\beta_1,\alpha_2,\beta_2)$ be the optimal policy.
The reward of pulling the first arm is $ R_1(\alpha_1,\beta_1)$.The reward of pulling the second arm is $ R_2(\alpha_2,\beta_2)$.
Since the success probability of arm1 has a Beta distribution, then $\hat\theta_1=E(\theta_1)=\frac{\alpha_1}{\alpha_1+\beta_1}$,and the expected failure probability will be$1-\frac{\alpha_1}{\alpha_1+\beta_1}=\frac{\beta_1}{\alpha_1
+\beta_1} $ .If pulling the $i$-th arm is successful, we update the parameters as $\alpha_i\rightarrow\alpha_i+1,\beta_i\rightarrow\beta_i$ and receive the corresponding reward $1+\gamma R(\alpha_i+1,\beta_i,\cdots)$,else,we update the parameters as $\alpha_i\rightarrow\alpha_i,\beta_i\rightarrow\beta_i+1$ and do not receive any reward $0+\gamma R(\alpha_i,\beta_i+1,\cdots)$.

According to LOTP:
\begin{align*}
R_1(\alpha_1,\beta_1) &= \frac{\alpha_1}{\alpha_1+\beta_1}\left[1+\gamma R(\alpha_1+1,\beta_1,\alpha_2,\beta_2)\right]+\frac{\beta_1}{\alpha_1+\beta_1}\left[\gamma R(\alpha_1,\beta_1+1,\alpha_2,\beta_2)\right] \\
Similarly:
R_2(\alpha_2,\beta_2) &= \frac{\alpha_2}{\alpha_2+\beta_2}\left[1+\gamma R(\alpha_1,\beta_1,\alpha_2+1,\beta_2)\right]+\frac{\beta_2}{\alpha_2+\beta_2}\left[\gamma R(\alpha_1,\beta_1,\alpha_2,\beta_2+1)\right]
\end{align*}
In each selection, we choose the arm with best reward:

$$R(\alpha_1,\beta_1,\alpha_2,\beta_2)=\max(R_1(\alpha_1,\beta_1),R_2(\alpha_2,\beta_2))$$
